```python
from nba_py import team
from nba_py import constants
import pandas as pd
```

```python
team_id = [i['id'] for i in constants.TEAMS.values()]
```

```python
team_name = constants.TEAMS.keys()
```

```python
#define fetctched columns
#from base data
base_col = ["GP","W","FG_PCT","FG3_PCT","FT_PCT","PF","TOV"]
GP,W,FG_PCT,FG3_PCT,FT_PCT,PF,TOV = [],[],[],[],[],[],[]
#from advance data
adv_col = ["OFF_RATING" , "DEF_RATING"]
OFF_RATING , DEF_RATING = [],[]
#from scoring data
sco_col = ['PCT_PTS_2PT','PCT_PTS_3PT','PCT_PTS_FT']
PCT_PTS_2PT,PCT_PTS_FT,PCT_PTS_3PT = [],[],[]

```

```python
# define fetched structure
col_dic = {}
for col in base_col+adv_col+sco_col+['clutch_type']:
    col_dic[col]=[]

```

```python
i = 0
n = len(team_name)
for i in range(0,n):
    f = team.TeamClutchSplits(team_id[i])
    #use last5min_plusminus_5point
    col_dic['clutch_type'] = 'last5min_plusminus_5point'
    for col in base_col:
        col_dic[col].append(f.last5min_plusminus_5point()[col].values[0])
    f = team.TeamClutchSplits(team_id[i],measure_type='Advanced')
    for col in adv_col:
        col_dic[col].append(f.last5min_plusminus_5point()[col].values[0])
    f = team.TeamClutchSplits(team_id[i],measure_type='Scoring')
    for col in sco_col:
        col_dic[col].append(f.last5min_plusminus_5point()[col].values[0])

    i+=1
    if i % 5 == 0:
        print (1.*i)/n
```

```python
col_dic
```

```python
col_dic['team']=team_name
```

```python
result = pd.DataFrame(col_dic,columns=['team']+base_col+adv_col+sco_col+['clutch_type'])
```
